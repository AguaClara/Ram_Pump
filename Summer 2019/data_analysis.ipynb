```
import aguaclara.research.procoda_parser as pp
import matplotlib.pyplot as plt
import numpy as np
from aguaclara.core.units import unit_registry as u
import aguaclara.core.constants as c
import aguaclara.core.physchem as pc
```

```
url1 = 'https://raw.githubusercontent.com/AguaClara/ram_pump/master/Summer%202019/Data%20Collection/7-2-2019_spring_K146_test3.xls'
pp.notes(url1)
```

```
start = 15217 #should be more than 'start'
end = 25704 #should be less than 'stop'

x = (pp.column_of_time(url1,start,end)).to(u.s)
pressure = pp.column_of_data(url1, start, 1, end, 'cm')
airchamber = pp.column_of_data(url1, start, 2, end, 'cm')


plt.clf()
plt.plot(x,pressure,'-', label = 'Waste Valve')
plt.plot(x,airchamber,'-', label='Air Chamber')
plt.xlabel('Time (s)')
plt.ylabel('Pressure (cm)')
plt.legend()
plt.show()
```
